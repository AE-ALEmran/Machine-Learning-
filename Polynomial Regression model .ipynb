```python
#importing Library
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing dataset
```

```python
dataset = pd.read_csv('Position.csv')
x= dataset.iloc[:,1:-1].values
y= dataset.iloc[:, -1].values
```

```python
#Training the linear Regression model on the whole data set
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
#Training the polinomial regression model 
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 10)
x_poly = poly_reg.fit_transform(x)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
#Visualising the Linear Regression results
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg.predict(x), color = 'blue' )
plt.title('Truth Vs Bluff(Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the POlynomial Regression results
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, lin_reg_2.predict(x_poly), color = 'blue' )
plt.title('Truth Vs Bluff(Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising the POlynomial Regression results(for higher resulation and smooth curve)
```

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue' )
plt.title('Truth Vs Bluff(Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

```python
#Predecting new result with Linear Regression
```

```python
lin_reg.predict([[6.5]])
```

```python
#Predecting new result with Polynomial Regression
```

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
