```python
#Importing the libraries
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
```

```python
dataset = pd.read_csv('Position.csv')
x= dataset.iloc[:,1:-1].values
y= dataset.iloc[:, -1].values
```

```python
#Training the Random forest model on the whole dataset
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(random_state =10)
regressor.fit(x,y)
```

```python
#Predicting a new result
```

```python
regressor.predict([[6.5]])
```

```python
#Visualising the decision tree results high resulation
```

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape((len(x_grid),1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue' )
plt.title('Truth Vs Bluff(Random Forest Reressor model)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
